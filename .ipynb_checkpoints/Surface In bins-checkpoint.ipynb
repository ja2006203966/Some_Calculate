```python
import numpy as np
# import scipy as sp
```

```python
def cal_angle(v0, v1, n):
    # reigon [0, 2pi)
    #np.arccos => [0, pi]
    v0 = unit_vector(v0)
    v1 = unit_vector(v1)
    n_u = unit_vector(n)
    cos = np.dot(v0, v1)
    sin = np.dot(np.cross(n_u, v0), v1)
    if sin>=0:
        return np.arccos(cos)
    return np.arccos(cos) + 2*(np.pi - np.arccos(cos))
    

def vector_length(v):
    return np.sqrt(np.dot(v, v))

def unit_vector(v):
    if (l := vector_length(v)) == 0:
        return v
    return v/l

def unit_vector_np(v):
    condi = ((l := np.sqrt(np.sum(v*v, axis = -1))) == 0).astype(np.int)
    l = l + condi
    return v/l

def encode(index):
    return "_".join([str(float(i)) for i in index])
def decode(idxStr):
    return [float(s) for s in idxStr.split('_')]

def edge_encode(index):
    return "@".join(sorted(index))

def edge_decode(s):
    return s.split("@")


#O(8)*O(log3)
def getPoles(center):
    return [[center[0] + dx, center[1] + dy, center[2] + dz] for dx in [-0.5, 0.5] for dy in [-0.5, 0.5] for dz in [-0.5, 0.5] ]

def cal_pts_on_surf(pt, n, A, B):
    if B.shape[0] == 0:
        return []
    # A: outside point
    # B: inside points
    n = unit_vector(n)
    AP = pt - A
    PB = B - pt
    AB = B - A
    # l = np.dot(AP, -n)
    l = np.sum(AP*(-n), axis = -1)
    l2 = np.expand_dims(np.sum(PB*(-n), axis = -1), axis = -1)
    return A + AB*np.abs(l/(l + l2) )
     

def cal_surface_in_cube(n, pt, center):
    n = unit_vector(n)
    poles = np.array(getPoles(center))
    pt = np.array(pt)
    in_out_np = np.sign(np.sum(n*(poles - pt), axis = -1))
    outside_condi = (in_out_np == 1)
    inside_condi = (in_out_np == -1)
    on_surf_condi = (in_out_np == 0)
    
    surf_pts = list(poles[on_surf_condi])
    for pt_i in poles[outside_condi]:
        neighbor_condi = np.sum((pt_i == poles).astype(np.int), axis = -1) == 2
        B = poles[neighbor_condi&inside_condi] 
        if B.shape[0] == 0:
            continue
        tmp_surf_pts = list(cal_pts_on_surf(pt, n, pt_i, B))
        surf_pts += tmp_surf_pts
    return surf_pts
    
def cal_surf_by_pts(n, center, pt, pts):
    if len(pts)<3:
        return set()
    pts = np.array(pts)
    surfs = set()
    n = unit_vector(n)
    cp = np.dot(pt - center, n)*n + center
    v0 = pts[0] - cp
    angles = np.array([cal_angle(v0, pt_i - cp, n) for i, pt_i in enumerate(pts)])
    index = np.argsort(angles)
    for i, idx in enumerate(index):
        idxStrs = [encode(cp), encode(pts[idx]), encode(pts[index[i-1]])]
        surfs |= set([edge_encode(idxStrs)])
    return surfs
    
```

```python
n = unit_vector([1, 1, 1])
# pt = np.array([0.75, 0.75, 0.75])

center = np.array([0.5, 0.5, 0.5])
# pt = np.array([1, 0.75, 0])
pt = np.array([1, 0.25, 0])

```

```python
poles = np.array(getPoles(center))
onsurf = cal_surface_in_cube(n, pt, center)
surfs = cal_surf_by_pts(n, center, pt, onsurf)
```

```python
import matplotlib.pyplot as plt

poles = np.array(getPoles(center))

fig, ax = plt.subplots(figsize = (8, 8))
ax = fig.add_subplot(projection='3d')


for s in surfs:
    pts = edge_decode(s)
    tmp = []
    for pt_str in pts:
        pt  = decode(pt_str)
        ax.scatter(pt[0], pt[1], pt[2], c = 'red', s = 50, marker = 'o')
        tmp.append(list(pt))
    tmp = np.array(tmp)
    ax.plot_trisurf(tmp[:, 0], tmp[:, 1], tmp[:, 2],
                    linewidth=0.2, antialiased=True, color = 'gray', alpha = 0.2)
    
for pt in poles:
    ax.scatter(pt[0], pt[1], pt[2], c = 'blue', s = 50, marker = 'o') 
    neighbor_condi = np.sum((pt == poles).astype(np.int), axis = -1) == 2
    nbs = poles[neighbor_condi]
    for nb in nbs:
        ax.plot([pt[0], nb[0]], [pt[1], nb[1]], [pt[2], nb[2]], linewidth = 3, c = 'black')
plt.show()
```
